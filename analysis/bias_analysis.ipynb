```python
import pandas as pd

baseline = pd.read_csv("/content/results - baseline.csv")
```

```python
# loading the original baseline results
baseline = baseline.rename(columns={"baseline": "response"})
baseline
```

```python
#loading the results generated post-RAG pipeline
mitigated = pd.read_csv("/content/merged_file (1).csv")
```

```python
mitigated
```

```python
# comparing the responses of our original results to the results post-RAG

comparison_df = pd.DataFrame({
    "baseline_response": baseline['response'],
    "mitigated_response": mitigated['response']
})

response_changes = comparison_df.groupby(['baseline_response', 'mitigated_response']).size().reset_index(name='count')

print("Response Changes Between baseline and mitigated:")
print(response_changes.sort_values(by='count', ascending=False))
```

```python
df1_match = baseline.rename(columns={'response': 'response_1'})
df2_match = mitigated.rename(columns={'response': 'response_2'})

merged_df_1_2 = pd.merge(df1_match, df2_match, on=['question'], how='inner')

merged_df_1_2['result'] = merged_df_1_2.apply(lambda row: row['response_1'] == row['response_2'], axis=1)
```

```python
merged_df_1_2
```

```python
result_counts_1_2 = merged_df_1_2['result'].value_counts()
result_counts_1_2
```

```python
question_order = (
    merged_df_1_2.groupby('question')['result']
    .apply(lambda x: (x == True).sum())
    .sort_values(ascending=False)
)
```
